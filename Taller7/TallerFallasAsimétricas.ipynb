# Taller Fallas Asimétricas

```python
import numpy as np
from math import pi
```

## Punto 1

```python
# Secuencia positiva - negativa
z10 = 0.2j+0.05j
z12 = 0.1j
z13 = 0.11j

z20 = 0.05j+0.2j
z21 = z12
z23 = 0.12j

z31 = z13 
z32 = z23

y10, y12, y13 = 1/z10, 1/z12, 1/z13
y20, y21, y23 = 1/z20, 1/z21, 1/z23
y30, y31, y32 = 0, 1/z31, 1/z32

Y1 = np.array([[y10+y12+y13, -y12, -y13],
              [-y21, y20+y21+y23, -y23],
              [-y31, -y32, y30+y31+y32]])

Z1 = np.linalg.inv(Y1)
Z2 = Z1

print(Z1)
```

```python
#Secuencia cero

z10 = 0.09j+0.06j+0.05j
z12 = 0.3j
z13 = 0.31j

z20 = 0.05j
z21 = z12
z23 = 0.32j

z31 = z13 
z32 = z23

y10, y12, y13 = 1/z10, 1/z12, 1/z13
y20, y21, y23 = 1/z20, 1/z21, 1/z23
y30, y31, y32 = 0, 1/z31, 1/z32

Y0 = np.array([[y10+y12+y13, -y12, -y13],
              [-y21, y20+y21+y23, -y23],
              [-y31, -y32, y30+y31+y32]])

Z0 = np.linalg.inv(Y0)

print(Z0)
```

```python
a = np.exp(1j*2*pi/3)

A = np.array([[1,1,1],
              [1, a**2, a],
              [1, a, a**2]])
```

## Punto 2

```python
def print_complex(V):
    print("{} ∠ {}°".format(abs(V),np.angle(V)*180/pi))
```

```python
# Falla línea-tierra nodo 3 Zf=0
I3_0 = 1/(Z1[2,2]+Z2[2,2]+Z0[2,2])
I3_1, I3_2 = I3_0, I3_0

I3_012 = np.array([I3_0, I3_1, I3_2])
print("I3_012")
for i in I3_012:
    print_complex(i)
print()

I3_abc = np.matmul(A,I3_012)
print("I3_abc")
for i in I3_abc:
    print_complex(i)
print()

V2_0 = 0-Z0[1,2]*I3_0
V2_1 = 1-Z1[1,2]*I3_1
V2_2 = 0-Z2[1,2]*I3_2

V2_012 = np.array([V2_0, V2_1, V2_2])
print("V2_012")
for i in V2_012:
    print_complex(i)
print()

V2_abc = np.matmul(A,V2_012)
print("V2_abc")
for i in V2_abc:
    print_complex(i)
print()

IG2_0 = (0 - V2_0)/(0.05j)
IG2_1 = (1 - V2_1)/(0.05j+0.2j)
IG2_2 = (0 - V2_2)/(0.05j+0.2j)

IG2_012 = np.array([IG2_0, IG2_1, IG2_2])
print("IG2_012")
for i in IG2_012:
    print_complex(i)
print()

IG2_abc = np.matmul(A,IG2_012)
print("IG2_abc")
for i in IG2_abc:
    print_complex(i)
print()
```

```python
# Falla línea-línea nodo 2 Zf=0.12j
Zf = 0.12j
I2_0 = 0
I2_1 = 1/(Z1[1,1]+Z2[1,1]+Zf)
I2_2 = -I2_1

I2_012 = np.array([I2_0, I2_1, I2_2])
print("I2_012")
for i in I2_012:
    print_complex(i)
print()

I2_abc = np.matmul(A,I2_012)
print("I2_abc")
for i in I2_abc:
    print_complex(i)
print()

V1_0 = 0-Z0[0,1]*I2_0
V1_1 = 1-Z1[0,1]*I2_1
V1_2 = 0-Z2[0,1]*I2_2

V1_012 = np.array([V1_0, V1_1, V1_2])
print("V1_012")
for i in V1_012:
    print_complex(i)
print()

V1_abc = np.matmul(A,V1_012)
print("V1_abc")
for i in V1_abc:
    print_complex(i)
print()
    
IG1_0 = (0 - V1_0)/(0.09j+0.06j+0.05j)
IG1_1 = (1 - V1_1)/(0.2j+0.05j)
IG1_2 = (0 - V1_2)/(0.2j+0.05j)

IG1_012 = np.array([IG1_0, IG1_1, IG1_2])
print("IG1_012")
for i in IG1_012:
    print_complex(i)
print()

IG1_abc = np.matmul(A,IG1_012)
print("IG1_abc")
for i in IG1_abc:
    print_complex(i)
print()
```
