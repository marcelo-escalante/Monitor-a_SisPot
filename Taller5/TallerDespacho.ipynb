# Taller Despacho Económico

```python
import numpy as np
```

```python
a1, b1, c1 = 480, 8.5, 0.036
a2, b2, c2 = 170, 9.2, 0.028
a3, b3, c3 = 700, 7.1, 0.021

a = [a1, a2, a3]
b = [b1, b2, b3]
c = [c1, c2, c3]

Pmin = [80, 110, 65]
Pmax = [450, 560, 395]

Pds = (305, 822, 1298)
```

## Punto 1

```python
def lambd(a,b,c,Pd):
    n = len(a)
    return (Pd+sum([b[i]/(2*c[i]) for i in range(n)]))/sum([1/(2*c[i]) for i in range(n)])

def P_opt(b,c,lamb):
    n = len(b)
    return [(lamb-b[i])/(2*c[i]) for i in range(n)]

def cost_gen(a,b,c,P):
    n = len(a)
    return [a[i]+b[i]*P[i]+c[i]*P[i]**2 for i in range(n)]

def dispatch(a,b,c,Pd):
    
    #Costo incremental lambda
    lamb = lambd(a,b,c,Pd)
    
    #Potencias de generación para costo óptimo
    P_optimum = P_opt(b,c,lamb)
    
    #Costo por generador
    cost_generator = cost_gen(a,b,c,P_optimum)
    
    #Costo total
    cost_tot = sum(cost_generator)
    
    print("Potencia demandada: ", Pd)
    print()
    print("Costo incremental")
    print(lamb)
    print()
    print("Despacho de potencia óptimo")
    print(P_optimum)
    print()
    print("Costo por generador")
    print(cost_generator)
    print()
    print("Costo total")
    print(cost_tot)
```

```python
# Demanda 305 MW
dispatch(a,b,c,305)
```

```python
# Demanda 822 MW
dispatch(a,b,c,822)
```

```python
# Demanda 1298 MW
dispatch(a,b,c,1298)
```

## Punto 2

```python
def dispatch_limits(a,b,c,Pd,Pmin,Pmax,lamb_init):
    i=0
    n=len(a)
    lamb = lamb_init
    P_op = [(lamb-b[i])/(2*c[i]) for i in range(n)] #
    limit_condition = [Pmin[i]<=P_op[i]<=Pmax[i] for i in range(n)]
    deltaP = Pd-sum(P_op)
    print("Potencia demandada: ", Pd)
    print()
    print(i)
    print(P_op)
    print(deltaP) 
    print(all(limit_condition))
    print(lamb)
    print()
    c_aux=[]
    if abs(deltaP)<=0.0001:
        for k in range(n):
            if limit_condition[k] == False:
                P_op[k] = Pmin[k] if P_op[k] < Pmin[k] else Pmax[k]
                limit_condition[k] = True
                c_aux.append(k)
        deltaP = Pd-sum(P_op)   
        i+=1
        print(i)
        print(P_op)
        print(deltaP) 
        print(all(limit_condition))
        print(lamb)
        print()

    while abs(deltaP)>0.0001:
        lamb += deltaP/sum([1/(2*c[k]) for k in range(n) if k not in c_aux])
        P_op = [(lamb-b[k])/(2*c[k]) if k not in c_aux else P_op[k] for k in range(n)] #
        limit_condition = [Pmin[i]<=P_op[i]<=Pmax[i] for i in range(n)]
        deltaP = Pd-sum(P_op)
        i+=1
        print(i)
        print(P_op)
        print(deltaP) 
        print(all(limit_condition))
        print(lamb)
        print()
        if abs(deltaP)<=0.0001:
            for k in range(n):
                if limit_condition[k] == False:
                    P_op[k] = Pmin[k] if P_op[k] < Pmin[k] else Pmax[k]
                    limit_condition[k] = True
                    c_aux.append(k)
            deltaP = Pd-sum(P_op) 
            i+=1
            print(i)
            print(P_op)
            print(deltaP) 
            print(all(limit_condition))
            print(lamb)
            print()
            
    cost_generator = cost_gen(a,b,c,P_op)
    costo_unitario = np.array(cost_generator)/np.array(P_op)
    cost_tot = sum(cost_generator)
    
    print("Resultados:")
    print("Potencia demandada: ", Pd)
    print()
    print("Límites de generación")
    print("P_min",Pmin)
    print("P_max",Pmax)
    print()
    print("Costo incremental")
    print(lamb)
    print()
    print("Despacho de potencia óptimo")
    print(P_op)
    print()
    print()
    print("Costo por generador")
    print(cost_generator)
    print()
    print("Costo unitario")
    print(costo_unitario)
    print()
    print("Costo total")
    print(cost_tot)
```

```python
# Demanda 305 MW
dispatch_limits(a, b, c, 305, Pmin, Pmax, 10)
```

```python
# Demanda 822 MW
dispatch_limits(a, b, c, 822, Pmin, Pmax, 10)
```

```python
# Demanda 1298 MW
dispatch_limits(a, b, c, 1298, Pmin, Pmax, 10)
```
